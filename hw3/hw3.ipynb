```python
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from matplotlib.pyplot import figure, draw, pause, gca
from scipy.ndimage.filters import convolve as filter2
import numpy as np
from typing import Tuple
%matplotlib inline
```

```python
img1 = io.imread("./lena.bmp").astype("float")
xaxis = img1.shape[0]
yaxis = img1.shape[1]
img2 = np.zeros((xaxis, yaxis))
img2[1:xaxis, 1:yaxis] = img1[0:xaxis-1, 0:yaxis-1];
```

```python
# lamda option: 0.1, 1, 10
# interation option: 8, 80, 800, 1800
lamda = 0.1
iteration = 8
```

```python
# Horn schunck
HSKERN = np.array([[1/12, 1/6, 1/12],[1/6, 0, 1/6], [1/12, 1/6, 1/12]], float)
X = np.array([[-1, 1], [-1, 1]])*0.25
Y = np.array([[-1, -1], [1, 1]])*0.25
T = np.ones((2, 2))*0.25 

img1 = img1.astype(np.float32)
img2 = img2.astype(np.float32)

# set up initial velocities
U = np.zeros([img1.shape[0], img1.shape[1]])
V = np.zeros([img1.shape[0], img1.shape[1]])

# Estimate derivatives
dx = filter2(img1, X) + filter2(img2, X)
dy = filter2(img1, Y) + filter2(img2, Y)
dt = filter2(img1, T) + filter2(img2, -T)

for _ in range(iteration):
    u = filter2(U, HSKERN)
    v = filter2(V, HSKERN)
    der = (dx*u + dy*v + dt) / (1+ lamda*(dx**2 + dy**2))
    U = u - dx * der
    V = v - dy * der
        
```

```python
scale=5
quivstep=20

ax = figure(figsize=(6,6)).gca()
ax.imshow(img2, cmap='gray')
for i in range(0, U.shape[0], quivstep):
    for j in range(0, V.shape[1], quivstep):
        ax.arrow(j, i, V[i, j]*scale, U[i, j]*scale,
                 head_width=3, head_length=1,width=0.0001)
ax.set_title("Lambda:{}".format(lamda))
draw()
#ax.savefig("Lambda:{}Iter:{}".format(lamda, iteration))
```
